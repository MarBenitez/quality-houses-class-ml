## Selecting high correlated variables

```python
corr = df_clustering.drop(['apartment_id', 'cluster'], axis=1).corrwith(df_clustering['cluster'], method='spearman').sort_values(ascending=False)
corr
```

```python
df_clustering.drop('apartment_id', axis=1).corr()
```

```python
corr_matrix_all = df_clustering.select_dtypes(include=['int64', 'float64']).corr()

mask = np.triu(np.ones_like(corr_matrix_all, dtype=bool))
corr_matrix_masked = corr_matrix_all.mask(mask)
fig = px.imshow(corr_matrix_masked, width=800, height=800, title='Pearson Correlation between Variables', color_continuous_scale='Viridis')
fig.show()
```

```python
corr = corr_matrix_masked.abs().unstack().sort_values(ascending=False).reset_index()
corr.rename(columns={'level_0': 'Variable 1', 'level_1': 'Variable 2', 0: 'Correlation'}, inplace=True)
corr = corr[corr['Variable 1'] != corr['Variable 2']]
corr.head(20)
```

## Discretization

```python
df_class_disc = df_class.copy()
```

When we tried to discretize some variables, we found that certain columns could not be discretized because their values were not strictly increasing. Therefore, we used different methods to fix these columns.

### Boxcox Transformation

```python
df_class_box =df_class_disc.copy()
```

```python
df_class_box['view_highways_mean'] = df_class_box['view_highways_mean'] + 1
```

```python
df_class_box['view_highways_box'], lambda_val = boxcox(df_class_box['view_highways_mean'])
```

```python
df_class_box.describe().T
```

```python
df_class_box.head()
```

### Min-Max Scaling

```python
df_class_minmax = df_class_disc.copy()
```

```python
min_max_scaler = MinMaxScaler()
df_class_minmax[['view_greenery_mean', 'view_ground_mean','view_highways_mean', 'view_pedestrians_mean','view_railway_tracks_mean', 'view_mountains_sum', 'mean_morning']] = min_max_scaler.fit_transform(df_class[['view_greenery_mean', 'view_ground_mean','view_highways_mean', 'view_pedestrians_mean','view_railway_tracks_mean', 'view_mountains_sum', 'mean_morning']])
df_class_minmax.head()
```

```python
df_class_minmax.describe().T
```

### Log Transformation

```python
df_class_log = df_class_disc.copy()
```

```python
df_class_log['view_highways_log'] = np.log1p(df_class_log['view_highways_sum'])
df_class_log['view_railway_tracks_log'] = np.log1p(df_class_log['view_railway_tracks_mean'])
df_class_log['view_mountains_log'] = np.log1p(df_class_log['view_mountains_sum'])
```

```python
df_class_log.drop(['view_highways_mean', 'view_railway_tracks_mean', 'view_mountains_sum'], axis=1, inplace=True)
```

```python
df_class_log.info()
```

```python
df_class_log.describe().T
```

Finally, none of these methods solved the problem, so we decided not to use the columns that couldn't be discretized. We also chose not to use the correlation matrix for variable selection, as we wanted to work with as many columns as possible to train the model. The final discretization is available in the [3-classification](https://github.com/MarBenitez/quality-houses-class-ml/blob/main/3_classification.ipynb) notebook.
